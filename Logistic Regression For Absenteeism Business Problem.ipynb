```python
# Import necessary libraries
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

# Create the CustomScaler class
class CustomScaler(BaseEstimator, TransformerMixin): 
    
    def __init__(self, columns, copy=True, with_mean=True, with_std=True):
        # Initialize the StandardScaler
        self.copy = copy
        self.with_mean = with_mean
        self.with_std = with_std
        self.scaler = StandardScaler(copy=self.copy, with_mean=self.with_mean, with_std=self.with_std)
        # Specify the columns to be scaled
        self.columns = columns
        self.mean_ = None
        self.var_ = None
    
    # The fit method
    def fit(self, X, y=None):
        # Fit the scaler on the specified columns
        self.scaler.fit(X[self.columns], y)
        # Capture mean and variance from the StandardScaler
        self.mean_ = self.scaler.mean_
        self.var_ = self.scaler.var_
        return self
    
    # The transform method
    def transform(self, X, y=None, copy=None):
        # Record the initial column order
        init_col_order = X.columns
        
        # Scale the specified columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        
        # Identify columns that were not scaled
        X_not_scaled = X.loc[:, ~X.columns.isin(self.columns)]
        
        # Concatenate scaled and non-scaled features and retain the original order
        return pd.concat([X_not_scaled, X_scaled], axis=1)[init_col_order]
```
